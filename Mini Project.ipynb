#### As you can see, these data cannot be used because of the "+" and "," signs, so it is better to convert them from str to int by removing these signs.

```python
# Drop rows with missing values
df['Content Rating'] = df['Content Rating'].dropna()

# Fill missing values with the most frequent value in each column
df['Content Rating'] = df['Content Rating'].fillna(df['Content Rating'].mode().iloc[0])
```

##### To use the size column, we must look at the unique data of this column

```python
df.Size.value_counts()
```

##### As you can see, we have 3 data models in this section
* 1: Varies with device which is str and it is better to remove this str by replacing the average value of the numbers
* 2: m or megabytes that must be converted to a number with the appropriate factor and removing the last m of this str
* 3: k or megabytes, which should be converted to a number with the appropriate factor and removing the last k of this str

```python
# Function to convert 'M' to numeric and multiply by 100, and remove 'k'
def convert_size(value):
    if value.endswith('M'):
        return pd.to_numeric(value[:-1], errors='ignore') * 1000
    elif value.endswith('k'):
        return pd.to_numeric(value[:-1], errors='ignore')
    else:
        return pd.to_numeric(value, errors='coerce')

# Apply the function to the 'Size' column
df['Size'] = df['Size'].apply(convert_size)

# Replace NaN values with the mean of the 'Size' column
mean_size = df['Size'].mean()
df['Size'].fillna(mean_size, inplace=True)
# To make the numbers smaller and based on megabytes, we divide all these numbers by 1000
df['Size']=df['Size'].divide(1000)
df['Size'] = df['Size'].round(2)
```
